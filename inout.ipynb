```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
# Define parameters
## ppf x^2 + g y^2 = b
b = 10000.0
g = 0.10
a = 0.50
max_x = np.sqrt(b)
x = np.arange(0.1, max_x, .1)
y = ((b - x**2)/g)**(.5)
ucra_x = x**(1-a)/(1-a)
ucra_y = y**(1-a)/(1-a)
ut = ucra_x + ucra_y
psi = g**(1/(a+1))
xcra_aw = psi*(b/(psi**2+g))**.5
xfree_aw = np.sqrt(b/2)
yfree_aw = ((b - xfree_aw**2)/g)**(.5)
xfreed_aw = np.sqrt(b*g/(1+g))
yfreed_aw = ((b - xfreed_aw**2)/g)**(.5)
p_aw = yfree_aw/xfree_aw
pd_aw = yfreed_aw/xfreed_aw
print(xcra_aw, xfree_aw, p_aw,xfreed_aw,pd_aw)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x, ut, 'b-')
ax2 = ax1.twinx()
ax2.plot(x,x+y)
fig.tight_layout()
```
