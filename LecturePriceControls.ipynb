- Why would a government decide to impose price controls?  Because of political pressure by either buyers that want a lower price  (e.g. rent control in New York City) or sellers that want a higher price (e.g. minimum wage).

- there are two kinds of price controls:
    - Price ceilings that do not allow the price to exceed a certain limit, and
    - price floors that do not allow the price to fall below a certain limit.

- We can measure the well-being of consumer and producer by
  computing their surplus. 
- The consumer curplus (CS) is the difference between their
  willigness to pay and the price they actually paid. 
- The producer surplus (PS) is the profit producers make. That is,
  it is the price minus the cost of production. 
  
- A price ceiling causes inefficiency (gains from trade not realized) in the following ways:
    - Deadweight loss: The loss in total surplus associated with the transactions that no longer occur due to the price control
    - Inefficient allocation to consumers: because consumers will get  the controlled resource through luck, some consumer that don't really need the resoucer will get it, and some that are anxious to get it won't. 
    - Wasted resources, i.e. the time lost for participating in the market
    - Inefficiently low quality: the landlord has no incentive to offer good quality since the rent is low and the tenants are queuing to get an apartment.
    - Black markets: give the landlord some extra chas under the table to get the apartment. 


```python
## Call libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Start the graph
fig1, ax = plt.subplots()
q = np.arange(0.,11, .5)
ps= 0 + q
pd = 10 -q
ax.plot(q,ps,q,pd)
plt.ylabel('price')
plt.xlabel('quantity')
price_ceiling = 2
plt.axhline(y= price_ceiling, hold=None)
plt.text(1, 6, r'CS')
plt.text(.5, 1, r'PS')
plt.text(3, 5, r'DWL =(')
plt.text(4, 2.5, r'price ceiling')
#plt.text(30, 62, r'#supply')
conds = ps<=price_ceiling
ax.fill_between(q, price_ceiling, ps, where=conds)
ax.fill_between(q, pd, price_ceiling, where=conds)
ax.set_ylim(0, 11)
```

- A price floor causes inefficiency (gains from trade not
  realized) in the following ways:
  - Deadweight loss
  - Inefficient allocation of sales among sellers. 
  - Wasted resources: throwing away the surplus goods
  - inefficiently high quality
  - illegal activity to bypass the price floor 

- So, why are there price controls? The basic answert is that some
  people benefit from these policies and they are powerful and vocal
  enough to convince governments to undertake them. 
  

```python
#Start the graph
fig2, ax2 = plt.subplots()
ax2.plot(q,ps,q,pd)
plt.ylabel('price')
plt.xlabel('quantity')
price_floor = 7
plt.axhline(y= price_floor, hold=None)
plt.text(1, 8, r'CS')
plt.text(1, 4, r'PS')
plt.text(3, 5, r'DWL =(')
plt.text(4, price_floor, r'price floor')
#plt.text(30, 62, r'#supply')
conds = pd>=price_floor
ax2.fill_between(q, price_floor, ps, where=conds)
ax2.fill_between(q, pd, price_floor, where=conds)
ax2.set_ylim(0, 11)
```

- A **quantity control**, or **quota** is an upper limit  on the quantity of some good that can be bought or sold. 
- The total amount of the good that can be legally transacted is
  the quota limit.
- A government may lmit the quantity of a good in a market by
  issuing *licenses*, such as the taxi medallion system in
  NY. 
- The interesting feature of this model is that the consumers are
  willing to pay the demand price (\$6) but the producers are
  receiving the supply price (\$4).
- The difference is the __oppurtinity cost__ of the
  producer from not renting out their licence for \$2.  
- Quotas are a very common phenomenon since the middle ages. Professional associations (aka guilds in the middle ages) that
  want to protect the interests of their members, issue licencses that
  a seller must possess in order to participate in a market. 
- Apart from taxis, such licenses may be required in markets such
  as truck owners, notaries, etc. 
- A quota causes (i) deadweight loss and (ii) incentives for
  illegal activities. 


```python
#Start the graph
fig3, ax3 = plt.subplots()
ax3.plot(q,ps,q,pd)
plt.ylabel('price')
plt.xlabel('quantity')
quota = 4
plt.axvline(x= quota, hold=None)
plt.text(4, 5, r'DWL')
plt.text(4, 8, r'quota')
conds = np.logical_and(q>=quota, pd>=ps)
#plt.text(30, 62, r'#supply')
ax3.fill_between(q, pd, ps, where=conds)
ax3.set_ylim(0, 11)
```

## The case of taxis

- Most major American cities have long used a system to limit the
  number of operating taxicabs, typically a medallion system: Drivers
  must own or rent a medallion to operate a taxi, and the city issues
  a fixed number of them. In NY, which established its medallion system in 1937, that number is 13 437. The number has risen only gradually since the late
  90s, even as the city's economy has boomed.
- Chicago medallion sale prices averaged \$ 298 000. 
- A seven-mile ride from the Loop to the University of Chicago in a medallion taxi costs about \$26, including tip. The same trip cost \$12.29  with UberX, the lowest-cost service option from Uber.
