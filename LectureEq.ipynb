## Preliminaries 

The market demand is characterized by a linear equation. At different prices, how many units does the market demand? 

\begin{equation}
q^d = 30 - p/3
\end{equation}

Similarly, the supply tell us at different prices how many units the market offers

\begin{equation}
q^s = -5 + p/2
\end{equation}

Note that in this the value of $q^s = -5$ at $p=0$ is meaningless, but helps us to draw the linear functions. 


## Market equilibrium 

But first, let's compute the equilibrium price and quantity. 

Set $q^d=q^s$, or 

\begin{equation}
-5 + p/2= 30 - p/3 
\end{equation}

Unknows go to the LHS (right hand side) and constants to the RHS, 

\begin{equation}
p/3 + p/2 = 30+5
\end{equation}

\begin{equation}
6 \times (p/3 + p/2) = 6 \times (30+5)
\end{equation}

\begin{equation}
5p = 210 \rightarrow p^* = 42
\end{equation}

Now, we use $p^*$ to find $q^*$. You can work with either $q^s$, or $q^d$ given that in equilibrium $q^s = q^d$. Let's work with $q^s = -5 + 42/2 = 16 = q^*$


## Graphs, CS and PS
We would like to draw our functions and compute CS and PS. Note that the equations given are quantities as a function of prices. For the purposes of producing graphs, we need to write the equations as prices in function of quantities. Recall that we want to draw prices in the y-axis. Thus,

\begin{equation}
p^d = 90 - 3q
\end{equation}

The y-intercept is 90, and the slope is 3. 

Similarly, for the supply we have 

\begin{equation}
p^s = 10 + 2p
\end{equation}
The y-intercept is 10, and the slope is 2. 



```python
## Call libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Start the graph
fig1, ax = plt.subplots()
q = np.arange(0.,40, 4)
ps= 10 + 2*q
pd = 90 -3*q
ax.plot(q,ps,q,pd)
plt.ylabel('price')
plt.xlabel('quantity')
plt.text(5, 60, r'CS')
plt.text(5, 30, r'PS')
plt.text(30, 1, r'#demand')
plt.text(30, 62, r'#supply')
conds = pd >= ps
ax.fill_between(q, ps, 42, where=conds)
ax.fill_between(q, pd, 42, where=conds)
ax.set_ylim(0, 100)
print("PS is", (42-10)*16)
print("CS is", (90-42)*16)
print("TS is", (90-42)*16+(42-10)*16)
```

The area above $p^*$ under the demand curve is the CS while the area above the supply curve is the PS. 

The PS is then computed as the difference between $p^*$ and the y-intercept times the quantity of equilibrium, or
\begin{equation}
PS = (42-10)\times 16 = 512
\end{equation}

The CS is computed as the difference between the y-intercept and $p^*$ times the quantity of equiligrium, or 
\begin{equation}
PS = (90-42)\times 16 = 768
\end{equation}

Therefore, the total surplus is $512+768=1280$
